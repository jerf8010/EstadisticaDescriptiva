# Regresi√≥n lineal de la diabetes

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
diabetes = datasets.load_diabetes()
type(diabetes.data)
```

```python
diabetes.data.shape
```

```python
x = diabetes.data[:, 2].reshape(-1,1)
y = diabetes.target
```

```python
regressor = linear_model.LinearRegression()
regressor.fit(x,y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

$$\tilde{y} = 949.43526038 x + 152.1334841628967 $$

```python
plt.scatter(x, y, color = 'black')
plt.plot(x, regressor.predict(x), color = 'red', linewidth = 3)
plt.show()
```

$$ MSE = \frac{\sum_{i=1}^n (y_i-\tilde{y}_i)^2}{n} $$

```python
mean_squared_error(y, regressor.predict(x))
```

```python
r2_score(y, regressor.predict(x))
```

## Dividiento el conjunto en testing y training

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,  train_size = 0.8)
```

```python
regressor2 = linear_model.LinearRegression()
regressor2.fit(x_train, y_train)
```

```python
y_pred = regressor2.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
plt.scatter(x_train, y_train, color = 'black')
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor2.predict(x_train), color = 'blue')
plt.show()
```
