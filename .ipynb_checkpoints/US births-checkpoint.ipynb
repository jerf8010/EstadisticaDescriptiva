# Análisis del naciemiento de bebés en US

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Exploración de Datos

```python
births = pd.read_csv("https://raw.githubusercontent.com/joanby/r-basic/master/data/us-births.csv")
births
```

```python
births.shape
```

```python
births.head()
```

```python
births.tail()
```

### Análisis de Datos

Resumen anual

```python
births.pivot_table("births", index = "year", columns = "gender", aggfunc = "sum")
```

```python
Obtenemos información por decadas
```

```python
births['decade'] = (births['year'] // 10) * 10
births.head()
```

```python
births.pivot_table('births', index = 'decade',columns = 'gender', aggfunc = 'sum')
```

Los nacimientos fueron a la alza, en los 2000's hubo una pequeña baja

### Visualización

```python
%matplotlib inline
```

```python
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize = (15,15))
births.pivot_table("births", index = "year", columns = "gender", aggfunc = "sum").plot()
plt.ylabel("Total de Nacimientos por Año")
plt.xlabel("Año")
plt.title("Total de Nacimientos por Año")
plt.show()
```

```python
quartiles = np.percentile(births['births'], [25,50,75])
quartiles
```

```python
med = quartiles[1]
IQR = quartiles[2] - quartiles[0]
sig = 0.75* IQR
```

### Filtrando los datos

```python
births = births.query("(births > @med - 5 * @sig) & (births < @med + 5 * @sig)")
births
```

```python
births.shape
```

```python
births['day'] = births['day'].astype(int)
```

```python
births.head()
```

## Cambiando el formato de la fechas

```python
births.index = pd.to_datetime(10000 * births.year + 100 * births.month + births.day, format = "%Y%m%d")
```

```python
births.head()
```

```python
births['dayofweek'] = births.index.dayofweek
```

```python
births.head()
```

```python
births.pivot_table('births', index = "dayofweek", columns = 'decade', aggfunc = 'mean')
```

```python
plt.figure(figsize = (15,15))
births.pivot_table('births', index = "dayofweek", columns = 'decade', aggfunc = 'mean').plot()

plt.gca().set_xticklabels(['', 'Lun', 'Mar', 'Mie', 'Jue', 'Vie', 'Sab', 'Dom'])
plt.title("Promedio de Nacimientos por dia de la semana, para cada decada")
plt.xlabel("Día de la semana")
plt.ylabel("Promedio de nacimientos por día")
plt.show()
```

Hay menos nacimientos los fines de semana

```python
births_by_date = births.pivot_table('births', index = [births.index.month, births.index.day], aggfunc = 'mean')
```

```python
births_by_date.head()
```

Agregando los datos de años bisiestos

```python
births_by_date.index = [pd.datetime(2020, month, day) 
                        for (month, day) in births_by_date.index]
```

```python
births_by_date.head()
```

```python
plt.figure(figsize = (12,4))
births_by_date.plot()
plt.show()
```

Parece que los nacimientos son inducidos y esto lo hacela gente para queno coincida con las fechas festivas
